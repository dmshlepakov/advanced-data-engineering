```python
# Import & Settings
import pandas as pd
import numpy as np
import os
from google.colab import drive
```

```python
# Ensure that Google Drive root (My Drive) is the current directory
root = os.getcwd()
filepath = 'My Drive'
cwd = os.path.join(root, filepath)
os.chdir(cwd)
print(os.getcwd())
```

```python
# Collect all Parquet files in dfiles list
dfiles = []
for file in os.listdir():
    if file.endswith(".parquet"):
        dfiles.append(file)
print('Parquet files in current directory: ', '\n', dfiles)
```

```python
# Pandas dataset for Addresses
addr = pd.read_parquet(dfiles[0], engine='pyarrow')
```
